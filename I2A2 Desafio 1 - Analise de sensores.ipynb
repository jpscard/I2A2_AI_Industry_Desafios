## *4. An√°lise explorat√≥ria dos dados*

### *4.1 Boxplot*

```python
boxplot = df.boxplot(column=['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8'], patch_artist=False)
```

### *4.2 Gr√°ficos de Segmento*

```python
df.plot(subplots =True, sharex = True, figsize = (10,13));
```

```python
for i in df.columns:
    plt.plot(df[i], label=str(i))
    plt.legend()
plt.xlabel('Sensores')
plt.ylabel('Medidas')
plt.show()
```

### *4.3 Matriz de correla√ß√£o*

```python
f, ax = plt.subplots(figsize = (10,6))
sns.heatmap(df.corr(), annot = True, fmt=".4f", linewidths=.6)
plt.show()
```
