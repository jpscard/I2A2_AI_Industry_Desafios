# 2) Converter texto para matriz

## 1 - Importação das bibliotecas necessárias para visualização de dados

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 2 - Definição da função para converter um arquivo de texto em um array

```python
def ConvertTxtToArray(path):
    # Abre o arquivo no caminho especificado com codificação UTF-8
    f = open(path, 'r', encoding='utf-8')
    # Lê todo o texto do arquivo
    text = f.read()
    # Substitui quebras de linha por espaços e divide o texto em palavras
    arrayText = text.replace("\n", " ").split(" ")
    # Retorna o array de palavras
    return arrayText
```

## 3 - Definição da função para filtrar e processar o array de texto

```python
def handleString(arrayText, filter):
    # Inicializa um array para armazenar os dados processados
    arrayData = []
    # Itera sobre cada palavra no array de texto
    for word in arrayText:
        if word:
            # Remove caracteres indesejados da palavra e converte para minúsculas
            filter_word = ''.join(w for w in word if w not in filter).lower()
            # Adiciona a palavra filtrada ao array de dados
            arrayData.append(filter_word)
    # Retorna o array de dados processados
    return arrayData
```

## 4 - Definição da função para analisar a ocorrência de palavras


```python
def analyzingWordOccurrences(filteredTextArray):
    # Inicializa um dicionário para armazenar a contagem de palavras
    wordDict = {}
    # Continua o loop enquanto houver palavras no array
    while len(filteredTextArray) > 0:
        word = filteredTextArray[0]
        j = 0
        count = 0
        # Conta ocorrências da palavra e remove do array
        while j < len(filteredTextArray):
            if word == filteredTextArray[j]:
                filteredTextArray.pop(j)
                count += 1
            else:
                j += 1
        # Adiciona a palavra e sua contagem ao dicionário
        wordDict[word] = count
    # Retorna o dicionário com as ocorrências das palavras
    return wordDict
```

## 5 - Definição da função para gerar um histograma de probabilidade


```python
def generateProbabilityHistogram(wordDict):
    # Seleciona as top 20 palavras mais frequentes
    top20 = {k: wordDict[k] for k in sorted(wordDict, key=wordDict.get, reverse=True)[:20]}

    # Calcula as probabilidades de ocorrência para cada palavra
    total_words = sum(wordDict.values())
    values = (np.array(list(top20.values())) / total_words) * 100

    # Configura e exibe o histograma
    plt.figure(figsize=(10, 8))
    plt.bar(top20.keys(), values)
    plt.title("Probabilidade de ocorrência de palavras do texto (Top 20)")
    plt.xlabel("Palavras")
    plt.ylabel("Probabilidade (%)")
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

## 6 - Definição da função para converter um array de texto de volta para texto e salvar em um arquivo


```python
def ConvertArrayToTxt(arrayText, nameFile):
    # Inicializa uma string para armazenar o texto reconstruído
    textRead = " ".join(arrayText)
    # Abre/cria o arquivo para escrita
    with open(f'texto lido - {nameFile}', 'w', encoding='utf-8') as file:
        # Escreve o texto no arquivo
        file.write(textRead)
```

## 7 - Ponto de entrada do programa quando executado como script

```python
if __name__ == '__main__':
    # Exemplo de uso das funções definidas acima
    print("Base Text - Matrix Representation I")
    arrayText = ConvertTxtToArray('Base Text - Matrix Representation I - v2.txt')
    filteredTextArray = handleString(arrayText, ".,;()")
    wordDict = analyzingWordOccurrences(filteredTextArray)
    generateProbabilityHistogram(wordDict)
    ConvertArrayToTxt(arrayText, 'Base Text - Matrix Representation I - v2.txt')

    print("Base Text - Matrix Representation II")
    arrayText = ConvertTxtToArray('Base Text - Matrix Representation II - v2.txt')
    filteredTextArray = handleString(arrayText, '.,;()?:""')
    wordDict = analyzingWordOccurrences(filteredTextArray)
    generateProbabilityHistogram(wordDict)
    ConvertArrayToTxt(arrayText,'Base Text - Matrix Representation II - v2.txt')
```
